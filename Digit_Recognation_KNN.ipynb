```python
from tensorflow import keras
from tensorflow.keras import datasets
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=datasets.mnist.load_data()
```

```python
len(x_test)
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
x_train[0]
```

```python
plt.matshow(x_train[0])
plt.show()
```

```python
x_trained = x_train.reshape(len(x_train), 28*28)
x_tested = x_test.reshape(len(x_test), 28*28)
```

```python
plt.matshow(x_trained[0].reshape(28,28))
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(x_trained, y_train)
```

```python
knn.score(x_tested,y_test)
```

```python
knn.predict(x_tested)
```
